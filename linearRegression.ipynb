```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from pylab import rcParams

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10,8
```

```python
rooms = np.random.randn(100,1)+3
rooms[1:10]
```

```python
price = 265 + 6*rooms + abs(np.random.randn(100,1))
price[0:10]
```

```python
plt.plot(rooms, price, 'r^')
plt.xlabel('No of rooms 2024 Average')
plt.ylabel('2024 Average Home Price in $1000s')
```

```python
X = rooms
y = price

LinReg = LinearRegression()
LinReg.fit(X,y)
print(LinReg.intercept_, LinReg.coef_)
```

let mention math behind the model 

Line equation ==> y= mX + C 

m = coeficient of equation or slope (6.03518703)
C = interceptor (265.7583793) 




```python
print(LinReg.score(X,y))  ##score funtion return coefficient of determination R^2 of the prediction
```
